```python
import pandas as pd


file_path = 'Earthquakes_database.csv'
earthquakes_df = pd.read_csv(file_path)

# Data svetis konvertireba  datetime formatshi
earthquakes_df['Date'] = pd.to_datetime(earthquakes_df['Date'], errors='coerce')

#  'Date' svetis danawileba shemdeg svetebad :'Year', 'Month', 'Day' 
earthquakes_df['Year'] = earthquakes_df['Date'].dt.year
earthquakes_df['Month'] = earthquakes_df['Date'].dt.month
earthquakes_df['Day'] = earthquakes_df['Date'].dt.day

#  'Date' svetis dropi
earthquakes_df.drop(columns=['Date'], inplace=True)

# 50% ze meti gamotovebuli mnishvnelobis mqone svetebis dropi
threshold = 0.5 * len(earthquakes_df)
earthquakes_df.dropna(axis=1, thresh=threshold, inplace=True)

# Im rigebis dadropva sadac critical svetebs aqvt gamotovebuli mnishvnelobebi
critical_columns = ['Latitude', 'Longitude', 'Magnitude']
earthquakes_df.dropna(subset=critical_columns, inplace=True)

# Dropis shemdeg indexebis dareseteba
earthquakes_df.reset_index(drop=True, inplace=True)


print(earthquakes_df.head())
```

```python
import pandas as pd


data = {
    'Latitude': [19.246, 1.863, -20.579, -59.076, 11.938],
    'Longitude': [145.616, 127.352, -173.972, -23.557, 126.427],
    'Magnitude': [4.5, 5.8, 6.2, 4.0, 7.1],
    'Depth': [131.6, 80.0, 20.0, 15.0, 15.0],
    'Year': [1965, 1965, 1965, 1965, 1965]
}
df = pd.DataFrame(data)


melted_df = pd.melt(df, id_vars=['Year'], value_vars=['Latitude', 'Longitude', 'Magnitude', 'Depth'], var_name='Variable', value_name='Value')
print("Melted DataFrame:")
print(melted_df)


pivot_table_df = pd.pivot_table(df, values='Magnitude', index='Year', columns='Depth', aggfunc='mean')
print("\nPivot Table DataFrame:")
print(pivot_table_df)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Magnitude'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Magnitude')
plt.title('Magnitude of Earthquakes Over Years')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import numpy as np

avg_magnitude_by_depth = df.groupby('Depth')['Magnitude'].mean()

plt.figure(figsize=(10, 6))
plt.bar(avg_magnitude_by_depth.index, avg_magnitude_by_depth.values, color='skyblue')
plt.xlabel('Depth')
plt.ylabel('Average Magnitude')
plt.title('Average Magnitude of Earthquakes by Depth')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Depth'], df['Magnitude'], color='green', alpha=0.5)
plt.xlabel('Depth')
plt.ylabel('Magnitude')
plt.title('Magnitude vs. Depth of Earthquakes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'Earthquakes_database.csv'
earthquakes_df = pd.read_csv(file_path)

#'Date' svetis konvertireba datetime formatshi
earthquakes_df['Date'] = pd.to_datetime(earthquakes_df['Date'], errors='coerce')


critical_columns = ['Date', 'Latitude', 'Longitude', 'Magnitude']
earthquakes_df.dropna(subset=critical_columns, inplace=True)


earthquakes_df['Year'] = earthquakes_df['Date'].dt.year
earthquakes_df['Month'] = earthquakes_df['Date'].dt.month


monthly_avg_magnitude = earthquakes_df.resample('M', on='Date')['Magnitude'].mean().reset_index()


print("Resampled Data:")
print(monthly_avg_magnitude.head())


plt.figure(figsize=(10, 6))
plt.plot(monthly_avg_magnitude['Date'], monthly_avg_magnitude['Magnitude'], marker='o', linestyle='-', color='b')
plt.xlabel('Date')
plt.ylabel('Average Magnitude')
plt.title('Monthly Average Magnitude of Earthquakes')
plt.grid(True)
plt.tight_layout()
plt.show()
```
